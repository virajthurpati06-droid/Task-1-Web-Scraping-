```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://exoplanetarchive.ipac.caltech.edu/"
```

```python
response = requests.get(url)
```

```python
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
title = soup.title.text
```

```python
print(title)
```

```python
paragraphs = soup.find_all("p")
```

```python
data_list = []
```

```python
for p in paragraphs:
    text = p.get_text().strip()
    if text != "":
        data_list.append(text)
```

```python
df = pd.DataFrame({
    "Extracted_Text": data_list
})
```

```python
df.head()
```

```python
df.to_csv("nasa_exoplanet_scraped_data.csv", index=False)
```
